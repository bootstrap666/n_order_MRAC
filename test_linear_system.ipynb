```python
import numpy as np
import matplotlib.pyplot as plt
from linear_system import linear_system
```

```python
fs = 8000
S = linear_system( b=np.array([1, 0]),
                   a=np.array([1,0,4]),
                   sampling_rate=fs)
S.print()
```

```python
# Avaliando a resposta ao degrau unit√°rio
t_end = 12
input_signal=np.ones(fs*t_end)
initial_conditions = np.zeros(2)
output, states = S.simulate(input_signal, initial_conditions)
```

```python
plt.rcParams['text.usetex'] = True
plt.rcParams.update({'font.size': 14})
plt.figure(figsize=(10, 6))
plt.plot(np.arange(0,t_end,1/fs), input_signal,'b',label=r'$x(t)$')
plt.plot(np.arange(0,t_end,1/fs), output,'r',label=r'$y(t)$')
plt.legend(loc='best')
#plt.plot([t_init, t_end], [bm/bp,bm/bp],'b:')
#plt.plot([t_init, t_end], [(am-ap)/bp,(am-ap)/bp],'r:')
plt.xlabel(r'$t$ (time)')
plt.ylabel(r'Signal values')
plt.title(r'Linear System')
plt.grid(True,'both')
plt.xlim([0, t_end])
plt.show()
```

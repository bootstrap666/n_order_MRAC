```python
import numpy as np
import matplotlib.pyplot as plt
from linear_system import linear_system
```

```python
Ts = 100E-6
fs = 1/Ts
L = 1E-3
C = 100E-6
R = 100

b = np.array([1.0/(L*C)])
a = np.array([1.0, 1.0/(R*C), 1.0/(L*C)])
S = linear_system( b=b,
                   a=a,
                   sampling_rate=fs)
S.print_zoh()
```

```python
# Avaliando a resposta ao degrau unit√°rio
t_start=0
t_end = 0.02
t_sim = np.arange(t_start,t_end,1/fs)
input_signal = 311*np.sin(2*np.pi*60*t_sim)
initial_conditions = np.array([0, 0])
output, states = S.simulate_zoh(input_signal, initial_conditions)
```

```python
plt.rcParams['text.usetex'] = True
plt.rcParams.update({'font.size': 14})
plt.figure(figsize=(10, 6))
plt.plot(np.arange(0,t_end,1/fs), input_signal,'b',label=r'$x(t)$')
plt.plot(np.arange(0,t_end,1/fs), output,'r',label=r'$y(t)$')
plt.legend(loc='best')
#plt.plot([t_init, t_end], [bm/bp,bm/bp],'b:')
#plt.plot([t_init, t_end], [(am-ap)/bp,(am-ap)/bp],'r:')
plt.xlabel(r'$t$ (time)')
plt.ylabel(r'Signal values')
plt.title(r'Linear System')
plt.grid(True,'both')
plt.xlim([0, t_end])
plt.show()
```
